```python
import pandas as pd
import numpy as np
pd.set_option('precision',13)
```

```python
dfBatch1 = pd.DataFrame(pd.read_excel('./GA_data.xlsx', header=14, sheet_name='Batch1'))
dfBatch4 = pd.DataFrame(pd.read_excel('./GA_data.xlsx', header=14, sheet_name='Batch4'))
dfBatch5 = pd.DataFrame(pd.read_excel('./GA_data.xlsx', header=14, sheet_name='Batch5'))
dfBatch6 = pd.DataFrame(pd.read_excel('./GA_data.xlsx', header=14, sheet_name='Batch6'))
dfBatch7 = pd.DataFrame(pd.read_excel('./GA_data.xlsx', header=14, sheet_name='Batch7'))
```

```python
df = pd.concat([dfBatch1, dfBatch4, dfBatch5, dfBatch6, dfBatch7], ignore_index=True, sort =False)
```

```python
df = df[['Page','Pageviews','Entrances']].copy()
df
```

```python
df['ID_GA'] = df['Page'].str.extract('(?:.*id)([0-9]+)(?:.html)', expand=False)
```

```python
df['ShortURL'] = "https://www.blick.ch/"+df['ID_GA']+'/'
df
```

```python
#Export
dfToCrawl = df[['ShortURL']].copy()
dfToCrawl.to_csv('dfToCrawl.csv', sep='\t')
```

```python
#Redirect Chains Import
dfCrawled = pd.DataFrame(pd.read_excel('./redirect_chains.xlsx', header=0, sheet_name='1 - Redirects Chain Report (Lis'))
dfCrawled
```

```python
dfCrawled = dfCrawled[['Address','Final Address','Final Indexability']].copy()
dfCrawled['ID_Crawl'] = dfCrawled['Final Address'].str.extract('(?:.*id)([0-9]+)(?:.html)', expand=False)
dfCrawled = dfCrawled[dfCrawled['Final Indexability'] == 'Indexable']
dfCrawled
```

```python
#Merge
dfMerged = pd.merge(df, dfCrawled, left_on='ID_GA', right_on='ID_Crawl', how='outer')


```

```python
dfMerged = dfMerged[dfMerged['ID_Crawl'].notnull()]
dfMerged = dfMerged[dfMerged['ID_GA'].notnull()]
dfMerged

```

```python
dfMerged = dfMerged.drop('Address',1)
dfMerged
```

```python
dfMerged['Folder'] = dfMerged['Final Address'].str.extract('http[s]:\/\/.*?\/(.*?)\/', expand=False)
dfMerged
```

```python
dfMerged.to_csv('theResult.csv', sep='\t')
```

```python
pivot = pd.pivot_table(dfMerged,index=['Folder'], values=['Pageviews','Entrances'], aggfunc='sum')
pivot.fillna(0, inplace=True)
pivot.reset_index(inplace=True)
pivot = pivot.sort_values('Pageviews', ascending=True)
pivot
```
